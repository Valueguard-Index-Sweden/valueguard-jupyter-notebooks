## Index values

```python
referenceValue="100"
referenceDate="2005-01-15"
minDate="2005-01-01"

r = requests.get("https://analys.valueguard.se/mimerweb/index/indexNormalized?iId=" + str(index_id) + "&referenceValue=" + referenceValue + "&referenceDate=" + referenceDate + "&minDate=" + minDate)
index_values_df = pd.DataFrame(r.json()['values'])
```

```python
index_values_df.plot(x="indexAverageDate", y=["indexValue"])
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = (30,10)
plt.show()
```
